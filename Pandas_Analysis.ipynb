```python
import pandas as pd
import numpy as np

data = {
    'transaction_id': range(1, 21),
    'date': pd.date_range('2024-10-01', periods=20, freq='D'),
    'region': ['North', 'South', 'East', 'West', None, 'North', 'South', None, 'East', 'West',
               'North', 'South', 'East', 'West', 'North', None, 'East', 'West', 'North', 'South'],
    'product_category': ['Electronics', 'Clothing', None, 'Books', 'Electronics', 'Home',
                         'Clothing', 'Books', 'Electronics', None, 'Home', 'Clothing',
                         'Books', 'Electronics', 'Home', 'Clothing', 'Books', 'Electronics',
                         'Home', 'Clothing'],
    'sales_amount': [1200, 450, 890, None, 1500, 670, None, 340, 2100, 780,
                     560, None, 420, 1800, 920, 510, 380, None, 1100, 640],
    'quantity': [2, 5, 3, 1, None, 4, 2, 3, 1, 5,
                 3, 2, None, 1, 4, 3, 2, 1, None, 4],
    'customer_age': [25, 34, None, 45, 29, None, 38, 52, 27, 41,
                     33, None, 48, 26, 35, 42, None, 31, 39, 44],
    'payment_method': ['Credit Card', 'UPI', 'Cash', 'Debit Card', 'Credit Card',
                       'UPI', 'Cash', None, 'Credit Card', 'UPI', 'Debit Card',
                       'Cash', 'Credit Card', None, 'UPI', 'Cash', 'Debit Card',
                       'Credit Card', 'UPI', None]
}

df = pd.DataFrame(data)
df
```

Region & Product_category: Filled with mode (most frequent value)
Sales_amount: Filling with median
Quantity: Filling using forward fill (ffill)
Customer_age: Filling with mean (rounded to integer)


```python
df["region"].fillna(df["region"].mode()[0], inplace=True)
df["product_category"].fillna(df["product_category"].mode()[0], inplace=True)
df["sales_amount"].fillna(df["sales_amount"].median(), inplace=True)
df["quantity"].fillna(df["quantity"].ffill(), inplace=True)
df["customer_age"].fillna(mean_customer_age, inplace=True)
```

Payment_method: Droping rows where payment_method is missing

```python
df = df.dropna(subset=["payment_method"])
df.drop(["level_0", "index"], axis=1)
```

Calculating the total sales by region

```python
df.groupby("region")["sales_amount"].sum()
```

Calculating average sales by product_category

```python
df.groupby("product_category")["sales_amount"].mean().round(2)
```

Grouping by both region and product_category and calculating total sales and quantity andthen use reset_index() to flatten

```python
df.groupby(["region","product_category"]).agg({"sales_amount":"sum", "quantity":"sum"}).reset_index()
```

Displaying the top 3 region-product combinations by sales

```python
df.groupby(["region","product_category"])["sales_amount"].sum().sort_values(ascending=False).head(3)
```

Creating a function sales_range() that returns max - min of sales

```python
def sales_range(x):
  return x.max(), x.min()

sales_range(df["sales_amount"])
```

Apply sales_range to find sales range for each region

```python
df.groupby("region")["sales_amount"].agg(sales_range)
```

Using .agg() to calculate multiple metrics by region:
sales_amount: sum, mean, max
quantity: sum, min

```python
df.groupby("region").agg({"sales_amount": ["sum", "mean", "max"],"quantity": ["sum", "min"]})
```
